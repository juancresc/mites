```python
clus_files = 'data/wheat/mites/'
outfile = 'data/wheat/mites_consensus.fasta'
```

```python
clusters = [f for f in os.listdir(clus_files) if os.path.isfile(os.path.join(clus_files, f))]
```

```python
buffer = []
mite_count = 0
for cluster in clusters:
    c_dir = clus_files + cluster 
    fasta_seq = SeqIO.parse(c_dir, 'fasta')
    found = False
    count = 0
    dismiss = False
    seqname = ''
    for record in fasta_seq:
        if len(str(record.seq)) > 1100:
            dismiss = True
            seqname = record.id
            break
        if not 'MITE_T' in record.id:
            found = True
            seqname = record.id
        count += 1
    if not found and not dismiss:
        seqname = 'MITE_' + str(mite_count) + ' ' + record.id
        print(cluster, count, 'new', seqname)
        mite_count += 1
    if found and not dismiss:
        print(cluster, count, seqname)
    if dismiss:
        print(cluster, count, 'dismiss', seqname)
    else:
        seq = SeqRecord(record.seq, id=seqname, description="_")
        buffer.append(seq)
```

```python
SeqIO.write(buffer, outfile, "fasta")
```
